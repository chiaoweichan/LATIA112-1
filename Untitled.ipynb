```python
import pandas as pd
import io

# Open the file in binary mode and read it
with open('111_student.csv', 'rb') as file:
    content = file.read().decode('utf-8', errors='ignore')

# Create a DataFrame from the content
df = pd.read_csv(io.StringIO(content))

data = pd.read_csv('111_student.csv')

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rc('font',family='Microsoft JhengHei')
```

```python
#1.分別計算就讀日間部、就讀夜間部的學生數量

#使用條件篩選日間部、夜間部
day = df[df['日間∕進修別'] == 'D 日']
night = df[df['日間∕進修別'] == 'N 職']


#計算日間部、夜間部分別的學生數量
bachelor_day_count1 = len(day)
bachelor_night_count2 = len(night)


#印出結果
print("就讀日間部的學生數量 :",bachelor_day_count1)
print("就讀夜間部的學生數量 :",bachelor_night_count2)



bachelor_day_count = bachelor_day_count1
bachelor_night_count = bachelor_night_count2

#學生數量數據
student_counts = [bachelor_day_count, bachelor_night_count]

#學生部門標籤
departments = ['日間部', '夜間部']

#繪製圓餅圖
plt.figure(figsize=(8, 8))
plt.pie(student_counts, labels=departments, autopct='%1.1f%%', startangle=140)
plt.axis('equal')#保持圓形
plt.title('日間部和夜間部學生分布')

#顯示圖表
plt.show()
```

```python
#2.碩士、學士、博士的數量
master = df[df['等級別'] == 'M 碩士']
bachelor = df[df['等級別'] == 'B 學士']
doctorate = df[df['等級別'] == 'D 博士']

#計算每個等級的數量
master_count = len(master)
bachelor_count = len(bachelor)
doctorate_count = len(doctorate)

#印出結果
print("碩士體系的數量:", master_count)
print("學士體系的數量:", bachelor_count)
print("博士體系的數量:", doctorate_count)


import matplotlib.pyplot as plt

master_count = master_count
bachelor_count = bachelor_count
doctorate_count = doctorate_count

#學生數量數據
student_counts = [master_count, bachelor_count, doctorate_count]

#學生等級標籤
levels = ['碩士', '學士', '博士']

#繪製圓餅圖
plt.figure(figsize=(8, 8))
plt.pie(student_counts, labels=levels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  
plt.title('碩士、學士和博士的數量分布')

#顯示圖表
plt.show()
```

```python
#3.計算所有大學(包括四技和五專)的男女比例

# 取得男生數量和女生數量
male = '男生計'  
selected_male = df[male]
column_sum_male = selected_male.sum()
female = '女生計'  
selected_female = df[female]
column_sum_female = selected_female.sum()

# 計算男女比例
total_students = column_sum_male + column_sum_female
male_percentage = (column_sum_male / total_students) * 100
female_percentage = (column_sum_female / total_students) * 100

# 將比例四捨五入到小數點第二位
male_percentage = round(male_percentage, 2)
female_percentage = round(female_percentage, 2)

print("男生數量：", column_sum_male)
print("女生數量：", column_sum_female)
print("男生比例：", male_percentage, "%")
print("女生比例：", female_percentage, "%")

import matplotlib.pyplot as plt

male_percentage = male_percentage
female_percentage = female_percentage

#學生性別比例數據
gender_percentages = [male_percentage, female_percentage]

#性別標籤
genders = ['男', '女']

#繪製圓餅圖
plt.figure(figsize=(8, 8))
plt.pie(gender_percentages, labels=genders, autopct='%1.1f%%', startangle=140)
plt.axis('equal') 
plt.title('所有大學的男女比例')

#顯示圖表
plt.show()
```

```python
#4.計算所有大學(包括四技和五專)的研修生男女比例

# 取得男生數量和女生數量
delaymale = '延修生男'  
selected_delaymale = df[delaymale]
column_sum_delaymale = selected_delaymale.sum()
delayfemale = '延修生女'  
selected_delayfemale = df[delayfemale]
column_sum_delayfemale = selected_delayfemale.sum()

# 計算男女比例
total_delaystudents = column_sum_delaymale + column_sum_delayfemale
delaymale_percentage = (column_sum_male / total_delaystudents) * 100
delayfemale_percentage = (column_sum_female / total_delaystudents) * 100

# 將比例四捨五入到小數點第二位
delaymale_percentage = round(delaymale_percentage, 2)
delayfemale_percentage = round(delayfemale_percentage, 2)

print("延修生男數量：", column_sum_delaymale)
print("延修生女數量：", column_sum_delayfemale)
print("延修生男比例：", delaymale_percentage, "%")
print("延修生女比例：", delayfemale_percentage, "%")

import matplotlib.pyplot as plt
import matplotlib.font_manager as fm


delaymale_percentage = delaymale_percentage
delayfemale_percentage = delayfemale_percentage

#學生性別比例數據
gender_percentages = [delaymale_percentage, delayfemale_percentage]

#性別標籤
genders = ['男', '女']

#繪製圓餅圖
plt.figure(figsize=(8, 8))
plt.pie(gender_percentages, labels=genders, autopct='%1.1f%%', startangle=140,)
plt.axis('equal')
plt.title('所有大學的延修生男女比例')
#顯示圖表
plt.show()
```

```python
#5.分別計算不同體系數量

#使用條件篩選一般 技職 師範
type1 = df[df['體系別'] == '1 一般']
type2 = df[df['體系別'] == '2 技職']
type3 = df[df['體系別'] == '3 師範']

#計算一般 技職 師範分別的數量
bachelor_type_count1 = len(type1)
bachelor_type_count2 = len(type2)
bachelor_type_count3 = len(type3)

#印出結果
print("一般體系的數量 :",bachelor_type_count1)
print("技職體系的數量 :",bachelor_type_count2)
print("師範體系的數量 :",bachelor_type_count3)

import matplotlib.pyplot as plt

bachelor_type_count1 = bachelor_type_count1
bachelor_type_count2 = bachelor_type_count2
bachelor_type_count3 = bachelor_type_count3

#不同體系的標籤
types = ['一般體系', '技職體系', '師範體系']

#對應的數量
type_counts = [bachelor_type_count1, bachelor_type_count2, bachelor_type_count3]

#繪製柱狀圖
plt.figure(figsize=(10, 6))
plt.bar(types, type_counts, color=['blue', 'green', 'red'])
plt.xlabel('不同體系')
plt.ylabel('學校數量')
plt.title('不同體系學校數量分布')

#顯示圖表
plt.show()
```

```python
#6.計算台大 師大 台科大學生數量

#計算台大學生數
tai_da_school_name = "國立臺灣大學"
tai_da_students = df[df['學校名稱'] == tai_da_school_name]['總計'].sum()

#計算師大學生數
shi_da_school_name = "國立臺灣師範大學"
shi_da_students = df[df['學校名稱'] == shi_da_school_name]['總計'].sum()

#計算臺科大學生數
taiko_da_school_name = "國立臺灣科技大學"
taiko_da_students = df[df['學校名稱'] == taiko_da_school_name]['總計'].sum()

#印出结果
print("台大的學生數：", tai_da_students)
print("師大的學生數：", shi_da_students)
print("台科大的學生數:",taiko_da_students)

import matplotlib.pyplot as plt

tai_da_students = tai_da_students
shi_da_students = shi_da_students
taiko_da_students = taiko_da_students

#學校名稱
schools = ['台大', '師大', '台科大']

#對應的學生數量
student_counts = [tai_da_students, shi_da_students, taiko_da_students]

#繪製柱狀圖
plt.figure(figsize=(10, 6))
plt.bar(schools, student_counts, color=['blue', 'green', 'red'])
plt.xlabel('學校')
plt.ylabel('學生數量')
plt.title('台大、師大和台科大的學生數量')

#顯示圖表
plt.show()
```

```python
#7.計算國立學校及非國立學校數量

#計算國立學校數量
national_school_count = len(df[df['學校名稱'].str.contains('國立')])

#計算非國立學校數量
non_national_school_count = len(df[~df['學校名稱'].str.contains('國立')])

#印出结果
print("國立學校數量：", national_school_count)
print("非國立學校數量：", non_national_school_count)

import matplotlib.pyplot as plt


national_school_count = national_school_count
non_national_school_count = non_national_school_count

#學校類型比例數據
school_percentages = [national_school_count, non_national_school_count]

#學校類型標籤
school_types = ['國立學校', '非國立學校']

#繪製圓餅圖
plt.figure(figsize=(8, 8))
plt.pie(school_percentages, labels=school_types, autopct='%1.1f%%', startangle=140)
plt.axis('equal') 
plt.title('國立學校和非國立學校的數量分布')

#顯示圖表
plt.show()
```

```python
#8.計算六都分別有幾所學校

# 去除重複的學校名稱並計算臺北市的學校數量
taipei_school_count = len(df[df['縣市名稱'] == '30 臺北市' ]['學校名稱'].unique())

# 去除重複的學校名稱並計算新北市的學校數量
newtaipei_school_count = len(df[df['縣市名稱'] == '01 新北市']['學校名稱'].unique())

# 去除重複的學校名稱並計算桃園市的學校數量
taoyuan_school_count = len(df[df['縣市名稱'] == '03 桃園市']['學校名稱'].unique())

# 去除重複的學校名稱並計算臺中市的學校數量
taichung_schools = df[(df['縣市名稱'] == '19 臺中市') | (df['縣市名稱'] == '06 臺中市')]
taichung_school_count = len(taichung_schools['學校名稱'].unique())

# 去除重複的學校名稱並計算高雄市的學校數量
kaohsiung_schools = df[(df['縣市名稱'] == '12 高雄市') | (df['縣市名稱'] == '50 高雄市')]
kaohsiung_school_count = len(kaohsiung_schools['學校名稱'].unique())

# 去除重複的學校名稱並計算台南市的學校數量
tainan_school_count = len(df[df['縣市名稱'] == '21 臺南市']['學校名稱'].unique())

# 打印结果
print("台北市的學校數量：", taipei_school_count)
print("新北市的學校數量：", newtaipei_school_count)
print("桃園市的學校數量：", taoyuan_school_count)
print("台中市的學校數量：", taichung_school_count)
print("高雄市的學校數量：", kaohsiung_school_count)
print("台南市的學校數量：", tainan_school_count)

import matplotlib.pyplot as plt

city_school_counts = [taipei_school_count, newtaipei_school_count, taoyuan_school_count, taichung_school_count, kaohsiung_school_count, tainan_school_count]

#六都標籤
cities = ['台北市', '新北市', '桃園市', '台中市', '高雄市', '台南市']

#繪製柱狀圖
plt.figure(figsize=(10, 6))
plt.bar(cities, city_school_counts, color=['blue', 'green', 'red', 'purple', 'orange', 'pink'])
plt.xlabel('城市')
plt.ylabel('學校數量')
plt.title('六都的學校數量')

#顯示圖表
plt.show()
```

```python
#9.計算師大的男女延修生數量，並計算男女比例

# 篩選師大的延修生
shi_da_school_name = "國立臺灣師範大學"
shi_da_extension_data = df[df['學校名稱'] == shi_da_school_name]

#計算師大延修生男女數量
male_extension_count = shi_da_extension_data['延修生男'].sum()
female_extension_count = shi_da_extension_data['延修生女'].sum()

#計算男女比例
total_extension_students = male_extension_count + female_extension_count
male_extension_percentage = (male_extension_count / total_extension_students) * 100
female_extension_percentage = (female_extension_count / total_extension_students) * 100

#將比例四捨五入到小數點後兩位
male_extension_percentage = round(male_extension_percentage, 2)
female_extension_percentage = round(female_extension_percentage, 2)

#印出结果
print("師大延修生男生數量：", male_extension_count)
print("師大延修生女生數量：", female_extension_count)
print("師大延修生男女比例：", male_extension_percentage, " %  :", female_extension_percentage, " % ")


import matplotlib.pyplot as plt

male_extension_count = male_extension_count
female_extension_count = female_extension_count
male_extension_percentage = male_extension_percentage
female_extension_percentage = female_extension_percentage

#延修生性別比例數據
extension_percentages = [male_extension_percentage, female_extension_percentage]

#性别標籤
genders = ['男', '女']

#繪製圓餅圖
plt.figure(figsize=(8, 8))
plt.pie(extension_percentages, labels=genders, autopct='%1.1f%%', startangle=140)
plt.axis('equal') 
plt.title('師大延修生男女比例')

#顯示圖表
plt.show()
```

```python
#10.排出人數最多前三名學校

#根據學校名稱分组並計算總計人數
school_grouped = df.groupby('學校名稱')['總計'].sum()

#依照總人數由高到低排列
sorted_schools = school_grouped.sort_values(ascending=False)

#選擇前三名學校
top_three_schools = sorted_schools.head(3)

#印出结果
print("人數最多的前三名學校:")
for school, total_students in top_three_schools.items():
    print(f"{school} 的總人數：{total_students}")

import matplotlib.pyplot as plt

top_three_schools = top_three_schools

# 學校名稱和對應的總人數
schools = top_three_schools.index
student_counts = top_three_schools.values

# 繪製柱狀圖
plt.figure(figsize=(10, 6))
plt.bar(schools, student_counts, color=['blue', 'green', 'red'])
plt.xlabel('學校名稱')
plt.ylabel('總人數')
plt.title('人數最多前三名學校')

#旋轉x軸標籤以防止重疊
plt.xticks(rotation=15)

#顯示圖表
plt.show()
```
